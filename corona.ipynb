```python
def konverzija_u_string(datum):
    string = ''
    if (datum.day < 10): string += '0'
    string += str(datum.day) + '/'
    if (datum.month < 10): string += '0'
    string += str(datum.month) + '/' + str(datum.year)
    return string
```

```python
def lista_dnevni_rast(drzava):
    lista = []
    i=-1
    for dnevni_rast in np.array(izdvoji_drzavu(drzava)['cases'])[::-1]:
        if (i==-1): lista.append(int(dnevni_rast))
        else: lista.append(int(dnevni_rast)+lista[i])
        i +=1
    return lista
```

```python
lista_dnevni_rast('Croatia')
```

```python
def prikaz_kontinuiranog_rasta(drzava):
    base =dtt.datetime(2019, 12, 31)
    dates2 = np.array([base + dtt.timedelta(days = i*7) for i in range (0, 16)])
    fig, ax = plt.subplots()
    ax.plot(np.array(izdvoji_drzavu(drzava)['dateRep'])[::-1], lista_dnevni_rast(drzava))
    plt.xticks(rotation=40, ha='right')
    i = 0
    ax.set_xticks(dates2)
    plt.grid(linestyle = ':')
    return fig
```
