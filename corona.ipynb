```python
def izdvoji_drzavu(drzava):
    return data[data['countriesAndTerritories'] == drzava]
```

```python
izdvoji_drzavu('Serbia')
```

```python
def ukupan_broj_slucaja_drzave(drzava):
    return izdvoji_drzavu(drzava)['cases'].sum()
```

```python
ukupan_broj_slucaja_drzave('Croatia')
```

```python
def izdvoji_dan(date):
    return data[data['dateRep'] == date]
```

```python
izdvoji_dan('09/04/2020')
```

```python
def broj_novih_slucajeva_na_dan(date):
    return izdvoji_dan(date)['cases'].sum()
```

```python
broj_novih_slucajeva_na_dan('09/04/2020')
```

```python
datumi = data['dateRep'].drop_duplicates()
```

```python
input = ['a', 'b']
#*input rastavlja, tj iterira po elementima liste (ali kao argument funkcije, ne inače)
```

```python
def najveci_dnevni_rast():
    return np.array([broj_novih_slucajeva_na_dan(date) for date in datumi]).max()
```

```python
najveci_dnevni_rast()
```

```python
def dnevni_podaci_drzave(drzava):
    return [[date for date in izdvoji_drzavu(drzava)['dateRep']][::-1], [int(pod) for pod in izdvoji_drzavu(drzava)['cases']][::-1]]
```

```python
dnevni_podaci_drzave('Croatia')
```

```python
def prikazi_stopu(drzava):
    fig, ax = plt.subplots()
    ax.plot(*dnevni_podaci_drzave(drzava))
    plt.xticks(rotation=40)
    i = 0
    for label in ax.get_xaxis().get_ticklabels():
        label.set_horizontalalignment('right')
        if (i%6 != 0):
            label.set_visible(False)
        i += 1
    return fig
```
