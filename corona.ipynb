```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.2, 1./10 
# A grid of time points (in days)
t = np.linspace(0, 160, 160)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, S/1000, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I/1000, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/1000, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

```python
def usporedba(lista, argument):
    brojevi = np.array([izdvoji_drzavu(drzava)[izdvoji_drzavu(drzava)['cases']>=0].shape[0] for drzava in lista])
    broj = brojevi.max()
    i = 0
    for temp in brojevi:
        if (broj == temp): break;
        i += 1
    datumi = pripremi_podatke(lista[i], argument)
    fig, ax = plt.subplots()
    duljina = datumi[0].shape[0]
    for j in range (0, len(lista)):
        nova = lista_dnevni_rast(lista[j], argument)[::-1]
        while (len(nova) < duljina): nova.append(0)
        ax.plot(datumi[0], nova[::-1]/(np.array(izdvoji_drzavu(lista[j])['popData2018'])[0]/100), label = lista[j])
    plt.xticks(rotation=40, ha='right')
    ax.set_xticks(datumi[1])
    ax.legend(loc = 'best')
    if (argument == 'cases'): 
        ax.set_title('Usporedba kretanja broja zaraženih')
        ax.set_ylabel('Udio zaraženih (%)')
    elif (argument == 'deaths'): 
        ax.set_title('Usporedba kretanja broja preminulih')
        ax.set_ylabel('Udio preminulih (%)')
    ax.set_xlabel('Datum (GGGG-MM-DD)')
    plt.grid(linestyle = ':')
    return fig
```

```python
import statsmodels.api as sm

def pomocni_ln(broj):
    if (broj == 0): return 0
    else: return float(np.log(broj))

lnn = np.vectorize(pomocni_ln, otypes=[np.float])

def vrati_procjenu(a, b):
    return lambda x: (np.e**a)*((np.e**b)**x)

def linearna_regresija(drzava, argument):
    dates = np.array(izdvoji_drzavu(drzava)['dateRep'])[::-1]
    broj_dana = 0
    while(np.array(izdvoji_drzavu(drzava)[izdvoji_drzavu(drzava)['dateRep']<=dates[broj_dana]][argument]).sum()==0):
        #print(np.array(izdvoji_drzavu(drzava)[izdvoji_drzavu(drzava)['dateRep']==dates[broj_dana]][argument])[0])
        broj_dana += 1
    if (broj_dana > 2): broj_dana -= 2
    brojac_dana = [i for i in range(0, dates.shape[0]-broj_dana)]
    brojac_dana2 = sm.add_constant(brojac_dana)
    zaraza = lnn(np.array(lista_dnevni_rast(drzava, argument))[broj_dana:])
    model = sm.OLS(zaraza, brojac_dana2)
    rez = model.fit()
    print(rez.summary())
    return [vrati_procjenu(rez.params[0], rez.params[1]), brojac_dana]
```

```python
def crtaj(drzava, argument, fja, broj):
    dates = np.array(izdvoji_drzavu(drzava)['dateRep'])[::-1]
    odakle = dates.shape[0]-len(broj)
    n = dates[odakle:].shape[0]//10
    base = dates[odakle]
    dates2 = np.array([base + np.timedelta64(i*n, 'D') for i in range (0, 12)])
    #također ne želimo da nam se prikaže apsolutno svaki datum jer je to nepregledno
    fig, ax = plt.subplots()
    ax.plot(dates[odakle:], lista_dnevni_rast(drzava, argument)[odakle:], label = 'Pravi brojevi')
    ax.plot(dates[odakle:], np.vectorize(fja)(broj), label = 'Procjena')
    plt.xticks(rotation=40, ha='right')
    ax.set_xticks(dates2)
    if (argument == 'cases'):
        ax.set_title('Usporedba stvarnih brojeva i modela u državi: ' + drzava)
        ax.set_ylabel('Broj zaraženih')
    elif (argument == 'deaths'):
        ax.set_title('Usporedba stvarnih brojeva i modela u državi: ' + drzava)
        ax.set_ylabel('Broj preminulih')
    ax.set_xlabel('Datum (GGGG-MM-DD)')
    ax.legend(loc = 'best')
    plt.grid(linestyle = ':')
    return fig
```

```python
def pomocni_ln(broj):
    if (broj == 0): return 0
    else: return float(np.log(broj))

lnn = np.vectorize(pomocni_ln, otypes=[np.float])

def vrati_procjenu(a, b):
    return lambda x: (np.e**a)*((np.e**b)**x)

def podatci_graf_regresija(drzava):
    dates = np.array(izdvoji_drzavu(drzava)['dateRep'])[::-1]
    broj_dana = 0
    while(np.array(izdvoji_drzavu(drzava)[izdvoji_drzavu(drzava)['dateRep']<=dates[broj_dana]]['cases']).sum()==0):
        #print(np.array(izdvoji_drzavu(drzava)[izdvoji_drzavu(drzava)['dateRep']==dates[broj_dana]][argument])[0])
        broj_dana += 1
    if (broj_dana > 2): broj_dana -= 2
    brojac_dana = [i for i in range(0, dates.shape[0]-broj_dana)]
    brojac_dana2 = sm.add_constant(brojac_dana)
    zaraza = lnn(np.array(lista_dnevni_rast(drzava, 'cases'))[broj_dana:])
    model = sm.OLS(zaraza, brojac_dana2)
    rez = model.fit()
    print(rez.summary())
    funkcija = vrati_procjenu(rez.params[0], rez.params[1])
    n = dates[broj_dana:].shape[0]//10
    base = dates[broj_dana]
    dates2 = np.array([base + np.timedelta64(i*n, 'D') for i in range (0, 12)])
    #također ne želimo da nam se prikaže apsolutno svaki datum jer je to nepregledno
    fig, ax = plt.subplots()
    ax.plot(dates[broj_dana:], lista_dnevni_rast(drzava, argument)[broj_dana:], label = 'Pravi brojevi')
    ax.plot(dates[broj_dana:], np.vectorize(funkcija)(brojac_dana), label = 'Procjena')
    plt.xticks(rotation=40, ha='right')
    ax.set_xticks(dates2)
    ax.set_title('Usporedba stvarnih brojeva i modela u državi: ' + drzava)
    ax.set_ylabel('Broj zaraženih')
    ax.set_xlabel('Datum (GGGG-MM-DD)')
    ax.legend(loc = 'best')
    plt.grid(linestyle = ':')
    return fig
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import ipywidgets as widgets
plt.rcParams['figure.max_open_warning'] = 500
%matplotlib agg

# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma, sigma = 0.2, 1./10, 1
# A grid of time points (in days)
t = np.linspace(0, 200, 200)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma, sigma):
    S, I, R = y
    dSdt = -beta * S * I / N + sigma * R
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I - sigma * R
    return dSdt, dIdt, dRdt

def pomoc(sigma):
    y0 = S0, I0, R0
    # Integrate the SIR equations over the time grid, t.
    ret = odeint(deriv, y0, t, args=(N, beta, gamma, sigma))
    S, I, R = ret.T

    # Plot the data on three separate curves for S(t), I(t) and R(t)
    fig, ax = plt.subplots()
    ax.plot(t, S/1000, 'b', alpha=0.4, lw=2, label='Susceptible')
    ax.plot(t, I/1000, 'r', alpha=0.4, lw=2, label='Infected')
    ax.plot(t, R/1000, 'g', alpha=0.4, lw=2, label='Recovered with immunity')
    ax.set_xlabel('Time /days')
    ax.set_ylabel('Number (1000s)')
    ax.set_ylim(0,1.2)
    ax.set_facecolor('#efefef')
    ax.yaxis.set_tick_params(length=0)
    ax.xaxis.set_tick_params(length=0)
    ax.grid(b=True, which='major', c='#fdfdfd', lw=2, ls='-')
    legend = ax.legend()
    legend.get_frame().set_alpha(0.5)
    for spine in ('top', 'right'):
        ax.spines[spine].set_visible(False)
    return fig

widgets.interact(pomoc, sigma = widgets.FloatSlider(value=0, min=0, max=0.2, step=0.005))
```
